# Create input file for sarek

```python
import pandas as pd
import json
```

```python
samples_dict = json.load(open('../../../../cases_ids.json','rb'))
samples_dict
```

```python
df = pd.DataFrame(columns=['patient','sex','sample','status','cram','crai'])
row = {}
for pt in samples_dict.keys():
    for t in ['tumor1','tumor2','normal']:
        
        row['patient'] = pt
        
        if samples_dict[pt]['sex'] == 'female':
            row['sex'] = 'XX'
        else:
            row['sex'] = 'XY'
            
        row['sample'] = samples_dict[pt][t]
        
        if t == 'normal':
            row['status'] = '0'
        else:
            row['status'] = '1'
            
        if pt == 'case4':
            row['cram'] = samples_dict[pt][t]+'.converted.cram'
            row['crai'] = samples_dict[pt][t]+'.converted.cram.crai'
        else:
            row['cram'] = samples_dict[pt][t]+'.recal.cram'
            row['crai'] = samples_dict[pt][t]+'.recal.cram.crai'
            
        df = df.append(row, ignore_index=True)
df
```

```python
df.to_csv('input_vc.csv',index=None)
```
